```python
import matplotlib.pyplot as plt
import cv2
import os
import time

import pandas as pd

dir = 'unrecognized'
names_list = []
labels_list = []
for f in [f for f in os.listdir(dir) if f != '.DS_Store']:
    names_list.append(f)
    img = cv2.imread(os.path.join(dir, f))
    plt.imshow(img)
    plt.show()
    time.sleep(1)


    label = input('enter the label:')
    labels_list.append(label)


pd.DataFrame({'filename': names_list, 'label': labels_list}).to_csv('labels.csv', index=False)
```
